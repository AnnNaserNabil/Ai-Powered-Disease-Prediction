```python
import pandas as pd
from fuzzywuzzy import process  # Or use RapidFuzz for faster results
from tqdm import tqdm  # Import tqdm for progress bar

# Load datasets
print("Loading datasets...")
symptom_disease_df = pd.read_csv("Testing.csv")  # Dataset 1: Symptoms -> Disease
disease_drug_df = pd.read_csv("final.csv")  # Dataset 2: Disease -> Drugs

# Function to find best matching disease name
def get_best_match(disease_name, choices):
    match, score = process.extractOne(disease_name, choices)  # Match from the disease-drug dataset
    return match if score > 80 else None  # Threshold to avoid wrong matches

# Apply mapping with progress tracking
print("Mapping diseases...")
tqdm.pandas(desc="Mapping diseases")
disease_drug_df["Mapped_Disease"] = disease_drug_df["disease"].progress_apply(
    lambda x: get_best_match(x, symptom_disease_df["prognosis"].unique())
)

print("Merging datasets...")
# Merge datasets
merged_df = symptom_disease_df.merge(disease_drug_df, left_on="prognosis", right_on="Mapped_Disease", how="left")

print("Saving merged dataset...")
# Save merged dataset
merged_df.to_csv("merged_dataset.csv", index=False)
print("Dataset saved successfully as merged_dataset.csv")

```
